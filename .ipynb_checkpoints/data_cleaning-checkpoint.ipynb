```python
# dependencies
import pandas as pd
```

```python
# get all the csv filepaths and create a list of files
bikedata_2020_01 = "data/JC-202001-citibike-tripdata.csv"
bikedata_2020_02 = "data/JC-202002-citibike-tripdata.csv"
bikedata_2020_03 = "data/JC-202003-citibike-tripdata.csv"
bikedata_2020_04 = "data/JC-202004-citibike-tripdata.csv"
bikedata_2020_05 = "data/JC-202005-citibike-tripdata.csv"
bikedata_2020_06 = "data/JC-202006-citibike-tripdata.csv"
bikedata_2020_07 = "data/JC-202007-citibike-tripdata.csv"
bikedata_2020_08 = "data/JC-202008-citibike-tripdata.csv"
bikedata_2020_09 = "data/JC-202009-citibike-tripdata.csv"
bikedata_2020_10 = "data/JC-202009-citibike-tripdata.csv"
bikedata_2020_11 = "data/JC-202009-citibike-tripdata.csv"
bikedata_2020_12 = "data/JC-202009-citibike-tripdata.csv"


all_files = [bikedata_2020_01,bikedata_2020_02,bikedata_2020_03,bikedata_2020_04,bikedata_2020_05,bikedata_2020_06,
             bikedata_2020_07, bikedata_2020_08, bikedata_2020_09, bikedata_2020_10, bikedata_2020_11, bikedata_2020_12]
```

```python
df = pd.concat((pd.read_csv(f, header = 0) for f in all_files))
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
# clean the df to have only the data with birth year over 1919 (under 100 years old)
df = df.loc[df["birth year"] > 1920, :]

# reset index
df = df.reset_index(drop = True)
```

```python
df.describe()
```

```python
# output the final dataframe to a csv file for using in tableau

df.to_csv("data/bikedata_2020.csv", index=False)
```
